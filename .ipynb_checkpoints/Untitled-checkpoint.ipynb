```python
import pandas as pd

file = "Clinic Sign-Up Prototype_2.xlsx"
df = pd.read_excel(file);
```

```python
def get_short(c): # Get short name of a clinic from a longer string
    if 'md' in c.lower(): return 'md';
    elif 'dermatology' in c.lower(): return 'dermatology';
    elif 'smoking' in c.lower(): return 'smoking';
    elif 'shelter' in c.lower(): return 'shelter';
    elif 'general' in c.lower(): return 'general';
    elif 'women' in c.lower(): return 'women';
    elif 'monday' in c.lower(): return 'monday';
    else: return None;
```

```python
def get_clinics_of_interest(long_name):
    _clinics = []
    for c in long_name.split(';'):
        _clinics.append(get_short(c));
    return _clinics
```

```python
users = []
for index, row in df.iterrows():
    user = {}
    user['id'] = row['ID']
    user['start_time'] = row['Start time']
    user['translator'] = row['Are you interested in being a translator (Spanish) instead of a general volunteer?']
    user['clinics_of_interest'] = get_clinics_of_interest(row['Please select any clinics you are interested in volunteering with'])
    user['name'] = row['First and Last Name']
    users.append(user)
```

## Results

```python
rdf = pd.read_excel(file, 'Results');
```

```python
clinics = {}
for index, row in rdf.iterrows():
    if pd.isnull(row['Clinic']): continue;
    date = row['Date']
    clinic_nameshort = get_short(row['Clinic'])
    
    info = {}
    info['clinic'] = clinic_nameshort
    info['volunteers'] = str(row['Volunteers']).split(',')
    
    if pd.isnull(row['Translators']):
        info['translators'] = str(row['Translators']).split(',')
    else:
        info['translators'] = []
    
    # existing:
    if clinic_nameshort in clinics.keys():
        clinics[clinic_nameshort][date] = info;
    # new:
    else:
        clinics[clinic_nameshort] = {}
        clinics[clinic_nameshort][date] = info;

clinics
```
