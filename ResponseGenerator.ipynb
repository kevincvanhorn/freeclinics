```python
import pandas as pd
import names
import random

N = 100
```

```python
trans_choices = ['Yes','No','I would like to be a general volunteer but can also speak Spanish.']
clinics = ['Agape MD Clinic','Agape Dermatology Clinic',
           'UGM Smoking Cessation Clinic', 'UGM Shelter Clinic',
           'BBHH General Clinic', 'BBHH Women\'s Clinic',
           'The Monday Clinic', 'BBHH Diabetes Clinic']

dMD = ['4/8 from 4:00-6:00 pm', '4/12 from 4:00-6:00 pm','4/16 from 4:00-6:00 pm','No Dates Available / Not Interested']
dDerm = ['4/15 from 7:00 am -12:00 pm', '4/29 from 7:00 am -12:00 pm', 'No Dates Available / Not Interested']
dSmoke = ['4/5 from 5:00-6:30 pm (Calvert)', '4/11 from 6-7:30 pm (CoH)',
         '4/12 from 5-6:30 (Calvert)', '4/19 from 5-6:30 (Calvert)',
         '4/25 from 6-7:30 (CoH)', '4/26 from 5-6:30 (Calvert)', 'No Dates Available / Not Interested']
dShelter = ['4/15 from 4:00-6:00 pm', '4/19 from 4:00-6:00 pm', 'No Dates Available / Not Interested'
           ]
dGen = ['4/13 from 5:45 -9:00 pm', '4/20 from 5:45 - 9:00 pm', 'No Dates Available / Not Interested']
dWomen = ['4/13 from 5:45-9:00 pm', '4/20 from 5:45-9:00 pm','No Dates Available / Not Interested']
dMon = ['4/16 from 4:00-6:00 pm','4/20 from 4:00-6:00 pm','No Dates Available / Not Interested']
dDia = ['4/16 from 4:00-6:00 pm','4/20 from 4:00-6:00 pm','No Dates Available / Not Interested']


def randomRanking():
    random.shuffle(clinics)
    return ';'.join(clinics)

def randomMulti(sample):
    i = random.randrange(len(sample))
    random.shuffle(sample)
    return ';'.join([sample[i] for _ in range(i)])
```

```python
form = {}
form['ID'] = list(range(N))
form['Start time'] = ['4/25/2023 22:03']*N
form['Completion time'] = ['4/25/2023 22:04']*N
form['Email'] = ['anonymous']*N
form['Name'] = ['']*N

form['First and Last Name'] = [names.get_full_name() for _ in range(N)]
form['UTSW Email Address'] = [names.get_last_name()+'@utsouthwestern.edu' for _ in range(N)]
form['What year are you in?'] = [random.choice(['Undergraduate','MS1','MS2','MS3','MS4']) for _ in range(N)]
form['Are you currently in the fourth-year elective?'] = [random.choice(['Yes','No']) for _ in range(N)]
form['Are you interested in being a translator (Spanish) instead of a general volunteer?'] = [random.choice(trans_choices) for _ in range(N)]
form['Please select any clinics you are interested in volunteering with'] = [randomMulti(clinics) for _ in range(N)]
form['Please rank your clinic preference'] = [randomRanking() for _  in range(N)]
form['Select Date Availability for Agape MD Clinic'] = [randomMulti(dMD) for _ in range(N)]
form['Select Date Availability for Agape Dermatology'] = [randomMulti(dDerm) for _ in range(N)]
form['Select Date Availability for UGM Smoking Cessation Clinic']= [randomMulti(dSmoke) for _ in range(N)]
form['Select Date Availability for UGM Shelter Clinic'] = [randomMulti(dShelter) for _ in range(N)]
form['Select Date Availability for BBHH General Clinic'] = [randomMulti(dGen) for _ in range(N)]
form['Select Date Availability for BBHH Women\'s Clinic'] = [randomMulti(dWomen) for _ in range(N)]
form['Select Date Availability for The BBHH Diabetes Clinic'] = [randomMulti(dDia) for _ in range(N)]
form['Select Date Availability for The Monday Clinic'] = [randomMulti(dMon) for _ in range(N)]
```

```python
df= pd.DataFrame.from_dict(form)
```

```python
df.to_excel('random_2023_04_v01.xlsx')
```
